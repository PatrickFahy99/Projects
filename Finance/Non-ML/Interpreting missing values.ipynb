# Interpreting missing values

Given a data set with n missing values represnted by "Missing_i" for i = 1,...,n estimate the values of these missing points.

```python
# Importing useful packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Collecting and formatting data 

data = pd.read_csv('Documents/Data.txt',sep = '\t', parse_dates = True)
df = pd.DataFrame(data)
df.columns = ['Price']
df.head()
```

```python
# Replace "Missing_i" with NaN for each missing entry

for i in range(len(df)):
    if df['Price'][i][0] == 'M':
        df['Price'][i] = np.NaN
```

```python
# Change entries from a string to a float

df['Price'] = df['Price'].astype(float)
```

```python
# Plot the data

df['Price'].plot()
```

```python
df2 = df.interpolate(method='time')

```

```python
# Plotting the estimated data

df2['Price'].plot()
```

```python
# Outputting the estimated values

for i in range(len(df)):
    if math.isnan(df['Price'][i]):
        print(df2['Price'][i])
```
