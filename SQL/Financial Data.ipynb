```python
import pandas as pd
import time
import datetime
from datetime import date
import pandas_datareader.data as web

symbols = ['^GSPC', '^DJI', 'FB', 'AMZN', 'TSLA', 'GOOGL', 'MSFT']

start = datetime.date(2007, 1, 1)

end = datetime.date.today()

data = {}
for i in symbols:
    try:
        data[i] = web.DataReader(i, 'yahoo', start, end)
        data[i].insert(0,'Symbol',i)
        print ('Done', i)
    except :
        print('No information for ticker: {}'.format(i))
        continue
```

```python
data[symbols[0]]
```

```python
import sqlite3

for i in range(len(symbols)):
    
    try:
        conn = sqlite3.connect('financialdata.db')
        data[symbols[i]].to_sql(symbols[i].replace('^', ''), conn, if_exists='replace', schema='financialdata.db', index=True)
    
    except:
        print('Error for {}'.format(symbols[i]))
        continue
```
