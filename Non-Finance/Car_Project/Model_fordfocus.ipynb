```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data_fordfocus')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
y = df['Price']
X = df.drop('Price', axis=1)
```

VERY low number of samples, so let's use a simple model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import KFold, cross_val_score

kfolds = KFold(n_splits=3, shuffle=True, random_state=42)

def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

def cv_rmse(model, X=X):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=kfolds))
    return (rmse)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import RidgeCV

ridge = make_pipeline(RobustScaler(), RidgeCV(cv=kfolds))
```

```python
from datetime import datetime

score = cv_rmse(ridge)
print("Regression: {:.4f} ({:.4f})\n".format(score.mean(), score.std()), datetime.now(), )
```

```python
from sklearn.metrics import mean_squared_error

ridge_model = ridge.fit(X_train, y_train)
print('RMSLE score on train data:')
print(rmsle(y, ridge_model.predict(X)))
```
