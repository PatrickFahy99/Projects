```python
from sklearn import preprocessing
```

```python
print(df.sqrt_price.mean(), df.sqrt_price.std())
print(df.sqrt_miles.mean(), df.sqrt_miles.std())
print(df.sqrt_age.mean(), df.sqrt_age.std())
print(df.log_lit.mean(), df.log_lit.std())
print(df.log_bhp.mean(), df.log_bhp.std())
```

```python
df['price_sc'] = preprocessing.scale(df.sqrt_price)
df['mile_sc'] = preprocessing.scale(df.sqrt_miles)
df['age_sc'] = preprocessing.scale(df.sqrt_age)
df['lit_sc'] = preprocessing.scale(df.log_lit)
df['bhp_sc'] = preprocessing.scale(df.log_bhp)
```

```python
df
```

```python
X = df.drop(['Price', 'Year', 'Miles', 'Litres', 'BHP', 'sqrt_price', 'sqrt_miles', 'age', 'sqrt_age', 'log_lit', 'log_bhp', 'price_sc'], axis=1)
```

```python
y = df['price_sc']
```

```python
X
```

```python
from sklearn.linear_model import ElasticNetCV, LassoCV, RidgeCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import LinearSVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
from mlxtend.regressor import StackingCVRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
```

```python
# Noticed this late - if doing this again i would group all hybrid together
```

```python
X
```

```python
validation_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=validation_size, random_state=42, shuffle=True)
```

```python
models = []
kfolds = 5

models.append(('RIDGE', RidgeCV(cv=kfolds)))
models.append(('LASSO', LassoCV(random_state=42, cv=kfolds)))
models.append(('ELASTIC NET', ElasticNetCV(cv=kfolds)))                               
models.append(('LinSVR', LinearSVR()))
models.append(('GBR', GradientBoostingRegressor(random_state =42)))
models.append(('XGBR', XGBRegressor(objective='reg:squarederror')))
```

```python
num_folds = 5

results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=42, shuffle=True)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='r2')
    results.append(cv_results)
    names.append(name)
    msg = "%s %f %f " % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# compare algorithms
fig = plt.figure()
fig.suptitle('Comparison of non-ensemble methods')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show();
```

```python
from sklearn.model_selection import GridSearchCV

param_test1 = {
 'max_depth':range(3,10,2),
 'min_child_weight':range(1,6,2)
}

gsearch1 = GridSearchCV(estimator = XGBRegressor( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'reg:squarederror', nthread=4, scale_pos_weight=1, seed=42), 
 param_grid = param_test1, scoring='r2',n_jobs=4,iid=False, cv=5, verbose=10)
gsearch1.fit(X_train,y_train)
```
